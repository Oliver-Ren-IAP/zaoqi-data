```python
import matplotlib.pyplot as plt 
import matplotlib as mpl
import numpy as np
from matplotlib.offsetbox import TextArea, DrawingArea, OffsetImage, AnnotationBbox
%config InlineBackend.figure_format = 'retina'

WRYH = mpl.font_manager.FontProperties(fname='WeiRuanYaHei-1.ttf') 

plt.rcParams['font.sans-serif'] = ['Songti SC']
```

```python
n = 10
x = [i for i in range(n)] # 共有多少个数据就设置多少个i
y1 = [1300] * n #这里是x轴最大范围，也就是浅色部分长度，需要根据你的数据进行修改
counts = [1222, 207, 112, 106, 76, 76, 65, 54, 51, 51] # 具体的每一个柱子长度
```

```python
plt.figure(figsize=(9, 6), dpi=150)

plt.xlim(0, 1310) # 这里的1310修改成最大的x轴，比上面的y1大一点即可

plt.barh(x, y1, color='#EDF6FF', alpha=0.99)

for i in range(n):
    
    
    plt.barh(x[i], counts[::-1][i], color='#00C5D2' if x[i] %
             2 == 0 else '#5172F0', height=0.4, ec='#333', lw=0.8)


ax = plt.gca()

ax.xaxis.set_ticks_position('top')
ax.tick_params(direction='in')

ax.tick_params(bottom=False, top=True, left=False, right=False)
plt.yticks([])

plt.ylim(-0.7, 9.7)


### 下面添加文字需要修改坐标，默认先不动直接执行，完了调整

plt.text(-125, 8.9, '  阿凡达  ', bbox=dict(boxstyle="round", fc="#EDF6FF",
                                         ec="black", alpha=0.9, linestyle='--', linewidth=0.8), size=9)
plt.text(-125, 7.9, '  Avatar  ', bbox=dict(boxstyle="round", fc="#EDF6FF",
                                            ec="black", alpha=0.9, linestyle='--', linewidth=0.8), size=9)
plt.text(-125, 6.9, '  潘多拉  ', bbox=dict(boxstyle="round", fc="#EDF6FF",
                                         ec="black", alpha=0.9, linestyle='--', linewidth=0.8), size=9)
plt.text(-105, 5.9, '  3D  ', bbox=dict(boxstyle="round", fc="#EDF6FF",
                                        ec="black", alpha=0.9, linestyle='--', linewidth=0.8), size=9)
plt.text(-115, 4.9, '  人类  ', bbox=dict(boxstyle="round", fc="#EDF6FF",
                                        ec="black", alpha=0.9, linestyle='--', linewidth=0.8), size=9)
plt.text(-115, 3.9, '  世界  ', bbox=dict(boxstyle="round", fc="#EDF6FF",
                                        ec="black", alpha=0.9, linestyle='--', linewidth=0.8), size=9)
plt.text(-125, 2.9, '  卡梅隆  ', bbox=dict(boxstyle="round", fc="#EDF6FF",
                                         ec="black", alpha=0.9, linestyle='--', linewidth=0.8), size=9)
plt.text(-115, 1.9, '  故事  ', bbox=dict(boxstyle="round", fc="#EDF6FF",
                                        ec="black", alpha=0.9, linestyle='--', linewidth=0.8), size=9)
plt.text(-115, 0.9, '  观影  ', bbox=dict(boxstyle="round", fc="#EDF6FF",
                                        ec="black", alpha=0.9, linestyle='--', linewidth=0.8), size=9)
plt.text(-120, -0.1, '  IMAX  ', bbox=dict(boxstyle="round", fc="#EDF6FF",
                                           ec="black", alpha=0.9, linestyle='--', linewidth=0.8), size=9)

for i in range(10):

    plt.text(counts[::-1][i]+14, i, f"{counts[::-1][i]}", fontsize=8, color='#00C5D2' if x[i] %
             2 == 0 else '#5172F0', verticalalignment='center',
             style='italic', family='serif', weight='heavy')

plt.text(0,10.5,'阿凡达影评都在说什么？',size = 20)

plt.text(800,2,'数据来源：豆瓣电影     采集时间：04/06/2021', verticalalignment='center',size =9,alpha = 0.7)
plt.text(800,1,'公众号：「早起Python」｜「可视化图鉴」', verticalalignment='center',size =9,alpha = 0.7)
    
plt.show()
```
